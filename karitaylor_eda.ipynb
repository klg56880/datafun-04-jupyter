## Import Dependencies and Data Aquisition 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load the Iris dataset into DataFrame
df = sns.load_dataset('iris')

# Inspect first rows of the DataFrame (to make sure the data has been loaded)
print(df.head())
```

## Initial Data Inspection

```python
print(df.head(10))
print(df.shape)
print(df.dtypes)
```

## Initial Descriptive Statistics 

```python
print(df.describe())
```

## Initial Data Distribution for Numerical Columns 

```python
# Inspect histogram by one numerical column
plt.hist(df['petal_length'], bins=20)

# Add Histogram Title
plt.title('Petal Length')

# Add X and Y Axis Labels
plt.xlabel('Length of Petals')
plt.ylabel('Number of Flowers')

# Command to Publish
plt.show()
```

##### **Observations**: The histogram above demonstrates the length of petals compared to total number of flowers. It's clear from the data that the majority of petal lengths fell between the 3 - 6 range; however, we also see a spike in number of flowers with petals lengths between 1 - 2.  

```python
data = pd.DataFrame({
    'data1': df['petal_length'],
    'data2': df['petal_width'],
    'data3': df['sepal_length'],
    'data4': df['sepal_width']
})

# Step 3: Create a figure and axes for the grid layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Step 4: Plot histograms on each axis
data['data1'].hist(ax=axs[0, 0], bins=20, color='blue', alpha=0.7)
axs[0, 0].set_title('Petal Length')

data['data2'].hist(ax=axs[0, 1], bins=20, color='orange', alpha=0.7)
axs[0, 1].set_title('Petal Width')

data['data3'].hist(ax=axs[1, 0], bins=20, color='green', alpha=0.7)
axs[1, 0].set_title('Sepal Length')

data['data4'].hist(ax=axs[1, 1], bins=20, color='red', alpha=0.7)
axs[1, 1].set_title('Sepal Width')

# Step 5: Customize layout
plt.tight_layout()

# Step 6: Display the grid of histograms
plt.show()
```

##### **Observations**: The grid of histograms above demonstrates the relationship that all 4 variables (petal length, petal width, sepal length, and sepal width) have to the total number of flowers. Now that we've looked at a basic analysis of the numerical data, we will now explore the categorical data from the Iris dataset. 

## Initial Data Distribution for Categorical Columns

```python
# Inspect value counts by categorical column
df['species'].value_counts()

# Inspect value counts for all categorical columns
for col in df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()
```

##### **Observations**: The histogram above shows all three different species of flowers included in the dataset and the total number of each. It is apparent that the dataset includes 3 equal subsets of flower species. 

## Initial Data Transformation and Feature Engineering

```python
sns.catplot(data=df, x="species", y="petal_length", kind="box")
```

```python
sns.catplot(data=df, x="species", y="petal_width", kind="box")
```

```python
sns.catplot(data=df, x="species", y="sepal_length", kind="box")
```

```python
sns.catplot(data=df, x="species", y="sepal_width", kind="box")
```

##### **Observation**: The box plots above illustrate the relationship between each of the 4 variables (petal length, petal width, sepal length, and sepal width) and all three flower species. The data shows that the Setosa species has the smallest petal length and width, as well as the smallest sepal length. On the other end of the spectrum, the Virginica species has the largest petal length, petal width, and sepal length. However, the last box plot tells a different story. It shows that the Setosa species has the greatest sepal width on average, followed by Virginica and Versicolor.
